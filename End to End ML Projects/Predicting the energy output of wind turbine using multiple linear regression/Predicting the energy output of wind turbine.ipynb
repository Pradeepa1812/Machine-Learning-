# Inorder to remove the 0 entries in each column

Since our dataset is large we can remove the rows with zero entries in each columns

```python
turbine_data=turbine_data[(turbine_data['Wind Speed (m/s)']>0) & (turbine_data['Theoretical_Power_Curve (KWh)']>0) & (turbine_data['Wind Direction (°)']>0) & (turbine_data['LV ActivePower (kW)']>0)] 
```

```python
#Inorder to verify any 0 entries in the dataset
turbine_data['Theoretical_Power_Curve (KWh)'].value_counts()
#so there is no zero entries present in the column and also for all the rest of the columns
```

# Implementation of machine learning algorithm

Inorder to perform machine learning algorithm we must want to assign dependent and independent attributes to a particular variables and here 'X' is our independent variable and 'y' is our dependent variable

```python
X=turbine_data[['Wind Speed (m/s)','Theoretical_Power_Curve (KWh)', 'Wind Direction (°)']] #independent variable
y=turbine_data[['LV ActivePower (kW)']] #dependent variable
```

# Train test split from scikit learn library

```python
from sklearn.model_selection import train_test_split
```

```python
#we split our dataset into training and test data for our machine learning model
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
```

and here X_train,X_test,y_train,y_test are two dimensional arrays

# Implementation of Linear Regression algorithm

```python
#importing Linear Regression class from the metrics package in scikit learn library
from sklearn.linear_model import LinearRegression
```

```python
#instantiating our model (Creation of an object)
lr=LinearRegression()
```

# Training our model / fitting training data into the model

```python
#training our machine learning model 
lr.fit(X_train,y_train)
```

```python
#assigning our independent value prediction to a another variable using predict method
y_pred=lr.predict(X_test)
```

# Calculating accuracy for our model 

Using score method

```python
print('Our model has an accuracy of : ',lr.score(X_test,y_test)*100,'%')
```

Visualizing using a scatter plot

```python
plt.scatter(y_test,y_pred)
plt.show()
#gives the best fitting line for our dataset
```

Visulaizing using an distance plot

```python
sns.distplot(y_test-y_pred,bins=100)
plt.show()
#There are appropriate mids and downs in our plot and hance it is a good model for prediction
```

using evaluation metrics from scikit learn package

i) Mean absolute error (MAE ):

```python
from sklearn.metrics import mean_absolute_error
print(mean_absolute_error(y_test,y_pred))
```

ii) Mean squared error (MSE) :

```python
from sklearn.metrics import mean_squared_error
print(mean_squared_error(y_test,y_pred))
```

iii)Root mean squared error (RMSE) :

```python
from sklearn.metrics import mean_squared_error
print(np.sqrt(mean_squared_error(y_test,y_pred)))
```
